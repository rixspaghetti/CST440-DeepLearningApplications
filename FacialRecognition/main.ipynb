#### Preprocessing

```python
import glob

### Check the structure of the dataset ###
# Find all images as well

# Define dataset path 
data_path = "helen_data"

# Find all image files 
image_files = glob.glob(os.path.join(data_path, "**", "*.jpg"), recursive=True)
print(f"Found {len(image_files)} image files.")

# Print a few image paths
image_files[:5]

### Resize and Normalize Images ### 

def preprocess_image(img_path, target_size=(256, 256)):
    # Load an image, resize it, and normalize pixel values 
    img = cv2.imread(img_path) # Read image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Convert from BGR to RGB
    img = cv2.resize(img, target_size, interpolation=cv2.INTER_LANCZOS4) # Resize image
    img = img / 255.0 # Normalize pixel values (0-1)
    return img

# Test on one image
sample_img = preprocess_image(image_files[0])

# Show preprocessed image 
plt.imshow(sample_img)
plt.axis('off')
plt.show()
```
